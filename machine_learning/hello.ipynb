```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
tips = sns.load_dataset("tips")

print(tips)
```

```python
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
plt.show()
```
